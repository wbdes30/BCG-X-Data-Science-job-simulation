## Data Visualization

#### Overall Churning Circumstances

```python
def annotation(x):
        if x == 1:
                SMALL_SIZE = 45
                MEDIUM_SIZE = 50
                BIGGER_SIZE = 55

                plt.rc('font', size=SMALL_SIZE)          # controls default text sizes
                plt.rc('axes', titlesize=SMALL_SIZE)     # fontsize of the axes title
                plt.rc('axes', labelsize=MEDIUM_SIZE)    # fontsize of the x and y labels
                plt.rc('xtick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
                plt.rc('ytick', labelsize=SMALL_SIZE)    # fontsize of the tick labels
                plt.rc('legend', fontsize=MEDIUM_SIZE)   # legend fontsize
                plt.rc('figure', titlesize=BIGGER_SIZE)  # fontsize of the figure title
        else:
                plt.rcdefaults()
```

```python
churn = len(client_df[client_df['churn'] == 1])
retention = len(client_df[client_df['churn'] == 0])

data = {'Churn': churn, 'Retention': retention}

labels = 'Churn', 'Retention' 
colors = ['silver', 'gold']
explode = (0.1, 0)

font = {'family': 'serif',
        'color':  'darkolivegreen',
        'weight': 'bold',
        'size': 50,
        }


plt.figure(figsize=(65, 55))
annotation(1)


plt.subplot(2, 2, 1)
# Churn status
plt.pie(data.values(), labels=labels, colors=colors, explode=explode, autopct='%1.1f%%', textprops={'fontsize': 50})
plt.title('Churning Status', fontdict=font)
plt.axis('equal')


plt.subplot(2, 2, 4)
# Churn by sale channels
sales_dat = client_df[['channel_sales', 'churn']]
sales_bar_data = sales_dat.groupby('channel_sales').sum().reset_index()
plt.bar(x=sales_bar_data.channel_sales, height= sales_bar_data.churn)
plt.xticks(rotation=30)
plt.xlabel('Sale channels')
plt.title("Churn by sale channels", fontdict=font)


plt.subplot(2,2,3)
# Churn by subscribed campaigns
origin_dat = client_df[['origin_up', 'churn']]
origin_bar_data = origin_dat.groupby('origin_up').sum().reset_index()
plt.bar(x=origin_bar_data.origin_up, height= origin_bar_data.churn, color = 'g')
plt.xticks(rotation=30)
plt.xlabel('Campaigns')
plt.title("Churn by subcribed campaigns", fontdict=font)


plt.subplot(2,2,2)
# Length of staying
client_df_copy = client_df
client_df_copy = client_df_copy.replace({'churn': {0: 'Retain', 1: 'Churn'}})
sns.boxplot(x='churn', y='num_years_antig', data=client_df_copy)
plt.xlabel('Churn status')
plt.ylabel('Year(s)')
plt.title('Customer Loyalty', fontdict=font)

plt.tight_layout()
plt.show()
```

Customer churn rate at PowerCo is around 10%, in which customers from `foosdfpfkusacimwkcsosbicdxkicaua` and `MISSING` sale channels or `lxidpiddsbxsbosboudacockeimpuepw` campaign accounted for the majority. In addition, many loyal customers are leaving PowerCo after many years of usage. As shown in the Customer Loyalty box plot, the median duration of retained customers using PowerCo service is 5 years, meanwhile many customers that has been loyal for over the same period of time are leaving. The interquartile ranges of staying time of both churn and retain customers fall between the same span of 4-6 years.

#### Correlation matrix

```python
from sklearn.preprocessing import LabelEncoder

corr_df = client_df
corr_df[['channel_sales', 'has_gas', 'origin_up']] = corr_df[['channel_sales', 'has_gas', 'origin_up']].apply(LabelEncoder().fit_transform)
```

```python
col_list = corr_df.select_dtypes(include='O').columns.to_list()
```

```python
# Correlation matrix

font = {'family': 'serif',
        'color':  'crimson',
        'weight': 'bold',
        'size': 23,
        }

corr_df = corr_df
for c in col_list:
    if c in corr_df.columns:
        corr_df.drop(columns=col_list, inplace=True)

corr_matrix = corr_df.corr()
plt.figure(figsize=(25, 25))
sns.heatmap(corr_matrix, annot=True, square=True, linewidths=0.5)
plt.title('Variables Relationship with Churn', fontdict=font, pad=30)
plt.show()
```

The correlation matrix indicates that there is no linear relationship between target variable `churn` and other features in the client_df dataset since the absolute values of their correlation coefficients are relatively close to 0.1 (from -0.074 to 0.096). This implies that the relationship between target variable and other predictors is non-linear, thus advanced regression models should be employed to capture the nonlinearity of the dataset and better predict the target variable.
